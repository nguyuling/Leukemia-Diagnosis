## Generate your own DNA sequences

**Required Information**
- Length of Sequence
- List of Nucleotides

```python
# define the length of the sequence
length=10

# supply the list of nucleotides/bases
bases=['A','C','G','T']
```

## Generate a Sequence

```python
# import random module
from random import choice
```

#### Approach 1 : using the for loop

```python
# creating an empty sequence
sequence=''

# randomly select a base and add it to the sequence (repeat 10 times)
for i in range(length):
    base=choice(bases)
    sequence+=base

print(sequence)
```

#### Approach 2 : using a list comprehension

```python
sequence=[choice(bases) for i in range(length)]
print(sequence)

# convert the list to a string
sequence=''.join(sequence)
print(sequence)
```

#### Approach 3 : using a while loop

```python
# create an empty list
sequence=''

while len(sequence)!=10:
    base=choice(bases)
    sequence+=base
    
print(sequence)
```

## Generate Multiple Sequences

#### Approach: Combine functions and loops

```python
# import the random module
from random import choice
```

```python
# create a function to generate a sequence
def generate_sequence(length=10):
    bases=['A','C','G','T']
    sequence=[choice(bases) for i in range(length)]
    sequence=''.join(sequence)
    return sequence
```

```python
# scenario 1: generate sequences of the same length

# use for loop to repeat the function to generate multiple sequences
length=10
number_of_sequences=5
sequences=[generate_sequence(length) for i in range(number_of_sequences)]
print(sequences)
```

```python
# scenario 2: generate sequences with different lengths

# user defined lengths
lengths=[10,5,6,5]
sequences=[generate_sequence(length) for length in lengths]
print(sequences)
```

```python
# computer generated random lengths and random sequence

from random import choice
from random import randint

# define the number of sequences
number_of_sequences = 5

# generate the lengths using randint (between 5 and 10)
lengths = [randint(5,10) for i in range(number_of_sequences)]
print(lengths)

sequences=[generate_sequence(length) for length in lengths]
print(sequences)
```

**`randint()` Syntax Explanation**

`randint(a, b)` generates a **random integer** between `a` and `b` (inclusive).

**Parameters:**
- `a` → lower bound (minimum value, included)
- `b` → upper bound (maximum value, included)

**Examples:**
- `randint(5, 10)` → randomly returns one of: 5, 6, 7, 8, 9, or 10
- `randint(1, 100)` → random integer from 1 to 100
- `randint(0, 1)` → random 0 or 1 (like a coin flip)

In the code above, `[randint(5,10) for i in range(number_of_sequences)]` creates a list of random sequence lengths, where each length is between 5 and 10 nucleotides.
