```python
sequence="AUGGGUAGCCCUAAUACUAGUCCUAAAAAG"

length=len(sequence)
length
```

```python
codons=[]
for i in range(0,30,3):
    codon=sequence[i:i+3]
    codons.append(codon)
    
print(codons)
print(len(codons))
```

```python
#replace all stop codons with '*'
sequence2="AUGUAGAGCCCUAAUACUAGUCCUAAAUGA"

codons=[]
for i in range(0,30,3):
    codon=sequence2[i:i+3]

    if codon=='UAG' or codon=='UAA' or codon=='UGA':
        codon='*'

    codons.append(codon)

print(codons)

codons_str = "-".join(codons)
print(codons_str)
```

```python
#create a function to get codons from a sequence

def get_codons(sequence):
    codons=[]
    length=len(sequence)
    for i in range(0,length,3):
        codon=sequence[i:i+3]
        codons.append(codon)
    return codons

get_codons("AUGGGUAGCCCUAAUACUAGUCCUAAAAAG")
```

```python
sequences=['AUGGUCAAA','AUUUGCCUA','UUCCAAGGG']

for sequence in sequences:
    codons=get_codons(sequence)
    print(codons)
```

#Exercise






*   Randomly generate a sequence of dna with length of 30.
*   Convert the DNA into RNA by replacing all occurrences of Thymine ('T') with Uracil ('U') while keeping other bases (Adenine 'A', Guanine 'G', Cytosine 'C') the same.
*   make a function to perform the convertion.


```python
from random import choice

# function to generate a random dna sequence
def generate_dna_sequence():
    length = 30
    bases = ['A', 'G', 'T', 'C']
    
    # randomly generate a dna sequence of given length
    dna = [choice(bases) for i in range(length)]
    
    # convert dna list to string and separate codons
    dna_str = "-".join(["".join(dna[i:i+3]) for i in range(0, length, 3)])
    return dna_str

# function to convert dna to rna
def convert_dna_to_rna(dna_str):
    rna_str = dna_str.replace('T', 'U')
    return rna_str

# function calling
dna_sequence = generate_dna_sequence()
rna_sequence = convert_dna_to_rna(dna_sequence)
print("DNA Sequence: ", dna_sequence)
print("RNA Sequence: ", rna_sequence)
```
