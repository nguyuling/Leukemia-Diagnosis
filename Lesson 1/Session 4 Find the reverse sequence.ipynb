## How to write the reverse of a sequence ###

### Case 1: When the sequence is a string

```python
# Use slicing
```

```python
DNA1="ATCTAGAGGATATAC"
```

```python
DNA1[::-1]
```

```python
reverse_sequence=DNA1[::-1]
```

**Syntax Explanation: `[::-1]`**

This is Python's **slice notation** with three components: `[start:stop:step]`

- `start` → (empty) = start from the beginning
- `stop` → (empty) = go to the end
- `step` → `-1` = move backwards by 1 position

So `[::-1]` means: "go through the entire string/list backwards, one character at a time"

**Examples:**
- `"ATCG"[::-1]` → `"GCTA"` (reversed)
- `[1,2,3,4][::-1]` → `[4,3,2,1]` (reversed)
- `"Hello"[::2]` → `"Hlo"` (every 2nd character)
- `"Hello"[1:4]` → `"ell"` (characters from index 1 to 3)

```python
print(reverse_sequence)
```

```python
print(DNA1)
print(reverse_sequence)
```

### Case 2: When dealing with a list data type

```python
# Approach 1
# convert list to a string and find reverse using slicing
```

```python
DNA2=['A','T','C','T','A','G','A','G','G','A','T','A','T','A','C']
```

```python
forward_sequence=''.join(DNA2)
```

```python
print(forward_sequence)
```

```python
reverse_sequence=forward_sequence[::-1]
```

```python
print(reverse_sequence)
```

```python
print(forward_sequence)
print(reverse_sequence)
```

```python
# Approach 3
# Apply reverse_method on the list and convert the list to a string
```

```python
DNA2=['A','T','C','T','A','G','A','G','G','A','T','A','T', 'A','C']
```

```python
# make a copy of the DNA2 list

sequence_copy=[]
sequence_copy.extend(DNA2)
```

```python
print(sequence_copy)
```

```python
# reverse method on a list
sequence_copy.reverse()
```

```python
print(sequence_copy)
```

```python
reverse_sequence=''.join(sequence_copy)
```

```python
print(reverse_sequence)
```

```python
print(''.join(DNA2))
print(reverse_sequence)
```
